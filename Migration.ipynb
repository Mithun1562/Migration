```python
import numpy as np
```

```python
years = np.arange(1980,2013)
immigrants = np.random.randint(2000,10000,33)
immigrants
```

```python
fig,ax = plt.subplots()
```

```python
ax.plot(years, immigrants)
plt.show()
```

```python
ax.plot(years, immigrants)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.hist(immigrants)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(years, immigrants,marker='s',color='green')

plt.title('Title')
ax.set_xlabel('Xlabel')
ax.legend(['Immigrants'],loc='upper right')
ax.set_xlim(1985,2012)
ax.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(years,immigrants,marker='o',markersize=3,linestyle = 'dotted',color='black')
plt.show()
```

```python
fig,ax =plt.subplots(1,2,sharey=True)
ax[0].plot(years, immigrants)
ax[1].scatter(years,immigrants)
plt.show()
```

```python
ax1=fig.add_subplot(1,2,1)
ax1.plot(years,immigrants)
ax2=fig.add_subplot(1,2,2)
ax2.scatter(years,immigrants)
plt.show()
```
